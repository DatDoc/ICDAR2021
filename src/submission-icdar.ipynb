```python
import os
import cv2 as cv
from imutils import paths
import numpy as np
```

```python
imagePaths = list(paths.list_images('/kaggle/input/test-data/TestCompetition_WITHOUTGroundTruth'))
```

```python
len(imagePaths)
```

```python
imagePaths.sort()
```

```python
imagePaths[0]
```

```python
predictions = []
```

```python
ls /kaggle/input/
```

```python
stt = 0
with open("/kaggle/input/d/phamlequangnhat/file-hand-printed/hand_printed.txt",'r') as f:
    for i in f.readlines():
        i = i.rstrip('\n')
        path = imagePaths[stt]
        if stt % 500 == 0:
            print('[INFO]', stt,'/55814')
        image = cv.imread(path)
        image = cv.resize(image,(224,224))
        x = np.expand_dims(image,axis=0)
        #result = style_model.predict(x,batch_size=1)
        
        if i == '0':
            answer = hand_model.predict(x,batch_size=1)
            predictions.append(np.argmax(answer))
        else:
            answer = printed_model.predict(x,batch_size=1)
            predictions.append(np.argmax(answer))
        stt += 1    
```

```python
len(predictions)
```

```python
fo = open('/kaggle/working/results_team.txt','w')
for i in predictions:
    fo.write(str(i)+'\n')
fo.close()
```
