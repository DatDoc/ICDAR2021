```python
import os
import cv2 as cv
from imutils import paths
import numpy as np
```

```python
imagePaths = list(paths.list_images('/kaggle/input/data-test/TestCompetition_WITHOUTGroundTruth'))
```

```python
imagePaths.sort()
```

```python
len(imagePaths)
```

```python
imagePaths[0]
```

```python
predictions = []
```

```python
stt = 0
for path in imagePaths:
    if stt % 500 == 0:
        print('[INFO]', stt,'/55814')
    image = cv.imread(path)
    image = cv.resize(image,(224,224))
    x = np.expand_dims(image,axis=0)
    #result = style_model.predict(x,batch_size=1)
    answer = model.predict(x,batch_size=1)
    predictions.append(np.argmax(answer))
    stt += 1
```

```python
len(predictions)
```

```python
fo = open('/kaggle/working/hand_printed.txt','w')
for i in predictions:
    fo.write(str(i)+'\n')
fo.close()
```
